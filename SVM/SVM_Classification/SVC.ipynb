```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_circles
import seaborn as sns
```

```python
x, y = make_circles(n_samples=500, noise=0.05, random_state=42)
```

```python
sns.displot(y)
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(x[y == 0, 0], x[y == 0, 1], c='blue', label='Class 0')
plt.scatter(x[y == 1, 0], x[y == 1, 1], c='red', label='Class 1')

plt.title('Circular Dataset')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_pred)
print(f"accuracy={acc}")
```

```python
x_min, x_max = x[:, 0].min() - 0.1, x[:, 0].max() + 0.1
y_min, y_max = x[:, 1].min() - 0.1, x[:, 1].max() + 0.1

xx, yy = np.meshgrid(np.linspace(x_min, x_max, 100), np.linspace(y_min, y_max, 100))
Z = lr.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, cmap=plt.cm.RdBu, alpha=0.8)
plt.scatter(x_test[y_pred == 0, 0], x_test[y_pred == 0, 1], c='blue', marker='x', label='Class 0 (Predicted)')
plt.scatter(x_test[y_pred == 1, 0], x_test[y_pred == 1, 1], c='red', marker='x', label='Class 1 (Predicted)')
plt.title('Decision Boundary and Predicted Labels')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
from sklearn.svm import SVC
svc=SVC()
svc.fit(x_train,y_train)
```

```python
y_pred2=svc.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
acc=accuracy_score(y_test,y_pred2)
print(f"accuracy={acc}")
```
