# Model Training

```python
x=Unlabeldata.values
```

```python
# scaler = StandardScaler()
x=power_transform(x)
```

```python
wcss=[]
for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit_predict(x)
    wcss.append(km.inertia_)
```

```python
plt.plot(range(1,11),wcss)
```

```python

km = KMeans(n_clusters=2)
labels = km.fit_predict(x)
```

```python
y=labelData['target'].values
```

```python
sns.displot(y)
```

```python
sns.displot(labels)

```

```python
t=0
total=len(y)
for i in range(len(y)):
    if labels[i]==y[i]:
        t+=1
    else:
        pass
```

```python
print(f"True predicted={t} out of {total}")
```
