```python
import seaborn as sns
from sklearn.datasets import load_breast_cancer
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.utils import resample
```

```python
data = load_breast_cancer()
```

```python
x = data.data
y = data.target
```

```python
sns.distplot(y)
```

```python
column_names = list(data.feature_names) + ['target']
df = pd.DataFrame(data=np.c_[x, y], columns=column_names)
```

```python
df.head()
```

```python
class_counts = df['target'].value_counts()
class_counts
```

```python
minority_class = class_counts.idxmin()
```

```python
df_majority = df[df['target'] == 1 - minority_class]
df_minority = df[df['target'] == minority_class]
```

```python
df_majority.head()
```

```python
df_minority.head()
```

```python
df_minority_upsampled = resample(df_minority, replace=True, n_samples=class_counts.max(), random_state=42)
```

```python
NewDf = pd.concat([df_majority, df_minority_upsampled])
NewDf = NewDf.sample(frac=1, random_state=42).reset_index(drop=True)
```

```python
sns.displot(NewDf['target'])
```

```python
NewDf.to_csv("labelData.csv",index=False)
```

```python
UnlableData=NewDf.drop(['target'],axis=1)
```

```python
UnlableData.to_csv("UnlabelData.csv",index=False)
```
