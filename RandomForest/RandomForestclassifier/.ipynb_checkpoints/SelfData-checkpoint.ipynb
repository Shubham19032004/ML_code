```python
from  sklearn.datasets import make_classification
from sklearn import tree
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score
```

```python
x, y = make_classification(
    n_samples=1000,  
    n_features=2,     
    n_informative=2,  
    n_redundant=0,    
    n_clusters_per_class=2, 
    flip_y=0.21,      
    random_state=42
)
```

```python
df = pd.DataFrame({'Feature1': x[:, 0], 'Feature2': x[:, 1], 'Target': y})
```

```python
df
```

```python
sns.scatterplot(x='Feature1', y='Feature2', hue='Target', data=df, palette='viridis')
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
precision=[]
max_depths = list(range(1, 100))
for i in range(1,100):
    tree=DecisionTreeClassifier(max_depth=i)
    tree.fit(x_train,y_train)
    y_pred=tree.predict(x_test)
    precision.append(precision_score(y_test, y_pred))      
```

```python
precision
```

### using GridSearchCV to find best parameters 

```python
param_grid = {
    'max_depth': [3, 5, 7, 10],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4],
    "criterion":['gini','entropy','log_loss']
}
```

```python
tree = DecisionTreeClassifier()
```

```python
#f1->f1_score
#precision->precision_score
grid_search=GridSearchCV(tree, param_grid, cv=5, scoring='precision')
```

```python
grid_search.fit(x_train, y_train)
```

```python
print("Best Parameters:", grid_search.best_params_)
print("Best Precision:", grid_search.best_score_)
```

```python
best_model = grid_search.best_estimator_
accuracy = best_model.score(x_test, y_test)
print(accuracy)
```

```python
_=tree.plot_tree(best_model, 
                   feature_names="features",  
                   class_names="0-1",
                   filled=True)
```
